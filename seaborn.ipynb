# seaborn

Categorical Data Plots
- bar plot
- countplot
- factorplot
- swarmplot
- violinplot
- stripplot

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# barplot and countplot

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
tips.info()
```

```python
sns.countplot(x='sex',data=tips)
```

# boxplot and violinplot

```python
sns.boxplot(x='day',y='total_bill', data=tips,palette='Set1')
```

```python
sns.boxplot(data=tips,palette='rainbow', orient = 'h')
```

```python
sns.boxplot(data=tips,palette='coolwarm', hue = 'smoker',x='day',y='tip')
```

```python
tips['percent'] = (tips['tip']/tips['total_bill'])*100
```

```python
tips.head(1)
```

```python
sns.boxplot(data=tips,palette='coolwarm', hue = 'sex',x='day',y='percent')
```

```python
sns.violinplot(x='day',y='percent',data=tips,palette='Set1',hue='sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',hue='sex',data=tips,palette='Set1',split=True)
```

# Combining categorical plots

```python
sns.violinplot(x='day',y='tip', data=tips, palette='rainbow')
sns.swarmplot(x='day', y='tip', data=tips, color='black', size=3)
```
